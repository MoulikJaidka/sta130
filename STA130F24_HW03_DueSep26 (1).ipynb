````python
import plotly.graph_objects as go
import seaborn as sns
import pandas as pd
import numpy as np

# Load penguins dataset from Seaborn
penguins = sns.load_dataset('penguins')

# Drop rows with missing flipper_length_mm values
penguins = penguins.dropna(subset=['flipper_length_mm'])

# Function to create histogram with statistical annotations
def create_histogram_with_stats(species_name, data):
    # Calculate statistical measures
    mean = data.mean()
    median = data.median()
    range_min = data.min()
    range_max = data.max()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    std = data.std()

    # Create histogram figure
    fig = go.Figure()

    # Add histogram trace
    fig.add_trace(go.Histogram(x=data, nbinsx=20, name=species_name, marker_color='lightblue', opacity=0.75))

    # Add vertical lines for mean and median
    fig.add_vline(x=mean, line_dash="dash", line_color="blue", annotation_text="Mean", annotation_position="top left")
    fig.add_vline(x=median, line_dash="dash", line_color="green", annotation_text="Median", annotation_position="top right")

    # Add shaded areas for range, interquartile range, and ±2 standard deviations
    fig.add_vrect(x0=range_min, x1=range_max, fillcolor="red", opacity=0.2, line_width=0, annotation_text="Range", annotation_position="bottom left")
    fig.add_vrect(x0=q1, x1=q3, fillcolor="yellow", opacity=0.2, line_width=0, annotation_text="IQR", annotation_position="bottom right")
    fig.add_vrect(x0=mean - 2*std, x1=mean + 2*std, fillcolor="purple", opacity=0.2, line_width=0, annotation_text="±2 Std", annotation_position="top right")

    # Update layout
    fig.update_layout(title=f'{species_name} Flipper Length Distribution',
                      xaxis_title='Flipper Length (mm)',
                      yaxis_title='Count')

    return fig

# Filter data for each species
adelie_data = penguins[penguins['species'] == 'Adelie']['flipper_length_mm']
chinstrap_data = penguins[penguins['species'] == 'Chinstrap']['flipper_length_mm']
gentoo_data = penguins[penguins['species'] == 'Gentoo']['flipper_length_mm']

# Create histograms for each species
fig_adelie = create_histogram_with_stats('Adelie', adelie_data)
fig_chinstrap = create_histogram_with_stats('Chinstrap', chinstrap_data)
fig_gentoo = create_histogram_with_stats('Gentoo', gentoo_data)

# Show each figure
fig_adelie.show()
fig_chinstrap.show()
fig_gentoo.show()

````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load penguins dataset from Seaborn
penguins = sns.load_dataset('penguins')

# Drop rows with missing flipper_length_mm values
penguins = penguins.dropna(subset=['flipper_length_mm'])
def create_kde_with_stats(ax, species_name, data):
    # Calculate statistical measures
    mean = data.mean()
    median = data.median()
    range_min = data.min()
    range_max = data.max()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    std = data.std()

    # Create KDE plot
    sns.kdeplot(data, ax=ax, fill=True, color="skyblue", label=species_name)
    
    # Add vertical lines for mean and median
    ax.axvline(mean, color="blue", linestyle="--", label="Mean")
    ax.axvline(median, color="green", linestyle="--", label="Median")
    
    # Add shaded areas for range, interquartile range, and ±2 standard deviations
    ax.axvspan(range_min, range_max, color="red", alpha=0.2, label="Range")
    ax.axvspan(q1, q3, color="yellow", alpha=0.2, label="IQR")
    ax.axvspan(mean - 2*std, mean + 2*std, color="purple", alpha=0.2, label="±2 Std")
    
    # Set titles and labels
    ax.set_title(f'{species_name} Flipper Length KDE')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')

    # Add legend
    ax.legend(loc="upper left")
# Filter data for each species
adelie_data = penguins[penguins['species'] == 'Adelie']['flipper_length_mm']
chinstrap_data = penguins[penguins['species'] == 'Chinstrap']['flipper_length_mm']
gentoo_data = penguins[penguins['species'] == 'Gentoo']['flipper_length_mm']

# Create a figure with subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Create KDE plots for each species
create_kde_with_stats(axes[0], 'Adelie', adelie_data)
create_kde_with_stats(axes[1], 'Chinstrap', chinstrap_data)
create_kde_with_stats(axes[2], 'Gentoo', gentoo_data)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
````

1. Box Plots
Description -
 A box plot (also called a whisker plot) summarizes the distribution of a dataset by showing the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The box represents the interquartile range (IQR). It's primarily used to show the spread and identify outliers. It gives a clear, quick summary of the dataset’s central tendency and variability.
 
PROS - 
Provides a concise summary of the data's distribution (median, IQR, range, outliers).
Great for comparing multiple distributions side-by-side.
Helps to quickly identify outliers and the spread of the data.

CONS - 
Lacks detail about the shape of the distribution (no insight into modality, skewness, or the exact density).
Can oversimplify complex distributions by hiding frequency details.

 
 2. Histograms
 Description - A histogram displays the distribution of a dataset by dividing the data into intervals (bins) and plotting the frequency of data points in each bin. It provides a clear view of how the data is distributed across different ranges.
 
 PROS - 
Easy to interpret; provides a clear picture of the frequency distribution.
Allows the viewer to see the shape of the distribution (skewness, modality, etc.).
Can show gaps, clusters, and outliers based on how the data is distributed in bins.

CONS - 
The appearance is sensitive to the bin width chosen (too few bins can hide details; too many bins can introduce noise).
Can appear "jagged" or uneven if the data set is small or the bin size is poorly selected.


3. Kernel Density Estimators
Description - KDE is a smooth, continuous estimate of the probability density function of a dataset. It’s essentially a smoothed version of a histogram, where each data point contributes to the density curve using a kernel function.

PROS - 
Provides a smooth and continuous estimate of the data distribution.
More accurate in representing the underlying distribution without relying on arbitrary binning.
Excellent for detecting multimodal distributions (i.e., more than one peak).

CONS -
Sensitive to bandwidth selection; too much smoothing can hide details, while too little can introduce noise.
May not represent actual frequencies as clearly as histograms (no clear count per range).
Not as concise as box plots for summary statistics.


I prefer Histograms

Rationale - 
1. Histograms provide a direct representation of the frequency of data points within specified intervals. This makes it easy to see how many observations fall within each range, allowing for quick assessments of common values and outliers. 
2. Histograms effectively show the overall shape of the data distribution, including its skewness and modality. 3. For larger datasets, histograms can summarize vast amounts of data efficiently, making it easier to spot trends, clusters, or gaps in the data.

In a perfectly symmetrical distribution (like a normal distribution), the mean and median are equal. Both measures of central tendency provide the same insight into the dataset.

Right-skewness- In a right-skewed distribution, the tail on the right side is longer or fatter than the left. Here, the mean is greater than the median because extreme values on the right (higher values) pull the mean up. The presence of higher values pulls the mean up, while the median remains closer to the center of the bulk of the data.

Left-skewness- In a left-skewed distribution, the opposite is true. The tail on the left side is longer, and the mean is less than the median because extreme values on the left (lower values) pull the mean down. The presence of lower values pulls the mean down, while the median remains higher due to the central values.

Additionally, we examined a code example that generated samples from right-skewed and left-skewed distributions, illustrating these concepts through histograms and calculations of mean and median. Overall, the discussion highlighted how understanding these relationships aids in accurately interpreting data distributions.

````python
# Importing required libraries
from scipy import stats
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

# Cell 1: Generating Right-Skewed Sample
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Cell 2: Plotting Histogram for Right-Skewed Sample
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Histogram of Right-Skewed Sample")
fig1.show(renderer="png")

# Cell 3: Calculating Mean and Median for Right-Skewed Sample
mean1 = sample1.mean()
median1 = np.quantile(sample1, [0.5])
mean1, median1  # Displaying mean and median

# Cell 4: Generating Left-Skewed Sample
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Cell 5: Plotting Histogram for Left-Skewed Sample
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Histogram of Left-Skewed Sample")
fig2.show(renderer="png")

# Cell 6: Calculating Mean and Median for Left-Skewed Sample
mean2 = sample2.mean()
median2 = np.quantile(sample2, [0.5])
mean2, median2  # Displaying mean and median
````

This relationship highlights how skewness affects the central tendency of a dataset, with the mean being sensitive to extreme values while the median remains more robust against them.
